The IMDb ID can be found in the movie's IMDb URL. For example, the ID for Inception (2010) is tt1375666 (https://www.imdb.com/title/tt1375666/). To run the code, press "Shift+Enter".
This example fetches the IMDb reviews page for the specified movie and uses BeautifulSoup to extract the reviews' titles, authors, dates, and content. It then uses IPython's display and Markdown functions to format and display the reviews in the Google Colab notebook.

Please note that web scraping IMDb might violate their terms of service. It's important to check IMDb's policies and ensure that you're allowed to perform web scraping on their content. Also, be aware that the code provided above may break if IMDb changes their website's structure.

````python
import requests
from bs4 import BeautifulSoup
from IPython.display import display, Markdown

def display_imdb_reviews(movie_id, num_reviews=5):
    url = f"https://www.imdb.com/title/{movie_id}/reviews"

    response = requests.get(url)
    if response.status_code != 200:
        print("Unable to fetch IMDb reviews")
        return

    soup = BeautifulSoup(response.text, "lxml")
    reviews = soup.find_all("div", class_="imdb-user-review")

    for idx, review in enumerate(reviews[:num_reviews]):
        title = review.find("a", class_="title").text.strip()
        author = review.find("span", class_="display-name-link").text.strip()
        date = review.find("span", class_="review-date").text.strip()
        content = review.find("div", class_="text").text.strip()

        display(Markdown(f"**Review {idx + 1}: {title}**"))
        display(Markdown(f"_by {author} on {date}_"))
        display(Markdown(f"{content}\n\n---"))

# Replace the movie_id below with the IMDb ID of the desired movie
movie_id = "tt1375666"  # Inception (2010)
display_imdb_reviews(movie_id)
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

random_numbers = np.random.randint(1, 1001, size=(1000, 1))

kmeans = KMeans(n_clusters=3)
kmeans.fit(random_numbers)

labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

plt.figure(figsize=(12, 6))
# Plot the data points
plt.scatter(range(len(random_numbers)), random_numbers, c=labels, cmap='viridis', s=30)

# Plot the cluster centers
plt.scatter([0]*len(cluster_centers), cluster_centers, c='red', marker='x', s=100, label='Cluster Centers')

plt.xlabel('Index')
plt.ylabel('Random Number')
plt.title('K-Means Clustering of 1000 Random Numbers')
plt.legend()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

random_numbers = np.random.randint(1, 1001, size=(1000, 1))

kmeans = KMeans(n_clusters=3)
kmeans.fit(random_numbers)

labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
plt.figure(figsize=(12, 6))
plt.scatter(range(len(random_numbers)), random_numbers, c=labels, cmap='viridis', s=30)

for center in cluster_centers:
    plt.axhline(y=center, color='red', linestyle='--', label=f'Cluster Center: {center[0]:.2f}')

# Remove duplicate labels in the legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

plt.xlabel('Index')
plt.ylabel('Random Number')
plt.title('K-Means Clustering of 1000 Random Numbers')
plt.show()
````
